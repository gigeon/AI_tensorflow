```python
t = np.arange(0,12,0.01)
y = np.cos(t)

plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('cos' 'graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('cos' 'graph')
plt.legend()
plt.grid()
plt.show()
```
